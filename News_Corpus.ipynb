```python
from nltk.corpus import stopwords
```

```python
nltk.download('all')  
```

```python
stop=stopwords.words("english")
```

```python
def stopwords_remove(text):
    # Appending words which are not stop words  
    removed= [i for i in text if i not in stop]  
    return removed
```

```python
stop_data=stopwords_remove(tokens)
```

```python
stop_data
```

```python
from nltk import Counter
```

```python
count = Counter(stop_data)
```

```python
word_freq = pd.DataFrame(count.items(),columns=['word','frequency']).sort_values(by='frequency',ascending=False)
```

```python
new_freq=word_freq[word_freq['frequency']>1]
```
